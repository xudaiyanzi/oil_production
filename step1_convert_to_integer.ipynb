```python
# there are two problems:

##     1 data types are differenct
#         need to convert the string into number for further analysis - to do prediction
    
##     2 it misses data
#         ~75% data in "breakdown pressure"
#         ~58% data in "water saturation"
#         ~15% data in "well spacing"
#         ~13% data in "proppant volume"
#         ~12% data in "porosity"
#         ~8% data in "isip"
#         ~5% data in "azimuth"
#         ~2% data in "tvd"
#         ~2% data in "toc"
#         ~2% data in "youngs modulus"
```

```python
#1 First convert data type into int

## 1.1 convert 'treatment company' into int

### 1.1.1 split the column of 'treatment company'
df[['treatment','company','tc_number']] = df['treatment company'].str.split(expand=True)

### 1.1.2 drop the column of 'treatment company','treatment','company', but keep 'tc_number' as 'treatment company'
df_drop_tc = df.drop(['treatment company','treatment','company'], axis=1)

# df_drop_tc.head()
```

```python
## 1.2 convert 'operator' into int

### 1.2.1 split the column of 'operator'
df_drop_tc[['operator_name','op_number']] = df_drop_tc['operator'].str.split(expand=True) 

### 1.2.2 drop the column of ''operator','operator_name', but keep 'op_number' as 'operator'
df_drop_op = df_drop_tc.drop(['operator','operator_name'], axis=1)

# df_drop_op.head()
```

```python
## 1.3 convert 'treatment company' and 'operator' into numbers

df_drop_op['tc_number'] = df_drop_op.tc_number.astype(int)
df_drop_op['op_number'] = df_drop_op.op_number.astype(int)

# df_drop_op.info()
```

```python
# 1.4 convert 'date on production' into int

## 1.4.1 convert 'date on production' into unix timestamp seconds

df_drop_op['date_sec'] = (pd.to_datetime(df_drop_op['date on production'])- pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')

# df_drop_op.info()
# df_drop_op['date_sec'].head()

## 1.4.2 drop 'date on production',keep 'date_sec' as 'date on production'

df_drop_dp = df_drop_op.drop(['date on production'], axis=1)

# finished converting the data type 

#df_drop_dp.head()
```

```python
# 1.5 finish the data type converting by saving it as a new dataframe
df_int = df_drop_dp
#df_int.info()
```

```python
cols = list(df_int.columns.values) #Make a list of all of the columns in the df
cols.pop(cols.index('production')) #Remove b from list

df_int = df_int[['production']+cols] #Create new dataframe with columns in the order you want
```
